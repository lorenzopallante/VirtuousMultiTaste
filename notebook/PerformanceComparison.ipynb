# Analyse results

```python
# analysed the predictions
bitter_sweet_pred = pd.read_csv(data_path + '/bittersweet/output.csv', sep=',')
bitter_sweet_pred[["name", "bitter_taste", "bitter_prob", "sweet_taste", "sweet_prob"]].head()
# if sweet_taste is True, then the compound is predicted as sweet
# if bitter_taste is True, then the compound is predicted as bitter
# if both are False, then the compound is predicted as other
bitter_sweet_pred["Predicted_Taste"] = np.where(bitter_sweet_pred["bitter_taste"] == True, "Bitter", np.where(bitter_sweet_pred["sweet_taste"] == True, "Sweet", "Other"))
# rename column name
bitter_sweet_pred[["name", "bitter_taste", "bitter_prob", "sweet_taste", "sweet_prob", "Predicted_Taste"]].head()
```

```python
# calculate performance
from sklearn.metrics import accuracy_score, f1_score, fbeta_score, precision_score, recall_score, auc   

test_performance = pd.DataFrame()
test_performance ['Set'] = ['BitterSweet']
test_performance ['Accuracy'] = [accuracy_score(bitter_sweet_pred["name"], bitter_sweet_pred["Predicted_Taste"]).round(4)*100]
test_performance ['F1'] = [f1_score(bitter_sweet_pred["name"], bitter_sweet_pred["Predicted_Taste"], average='weighted').round(4)*100]
test_performance ['F2'] = [fbeta_score(bitter_sweet_pred["name"], bitter_sweet_pred["Predicted_Taste"], average='weighted', beta=2).round(4)*100]
test_performance ['Precision'] = [precision_score(bitter_sweet_pred["name"], bitter_sweet_pred["Predicted_Taste"], average='weighted').round(4)*100]
test_performance ['Recall'] = [recall_score(bitter_sweet_pred["name"], bitter_sweet_pred["Predicted_Taste"], average='weighted').round(4)*100]
test_performance
```

```python
# create dataframe for sweet (all non sweet are other)
sweet_df = bitter_sweet_pred.copy()
sweet_df["Predicted_Taste"] = np.where(sweet_df["Predicted_Taste"] == "Bitter", "Other", sweet_df["Predicted_Taste"])
sweet_df["name"] = np.where(sweet_df["name"] == "Bitter", "Other", sweet_df["name"])

# create for bitter
bitter_df = bitter_sweet_pred.copy()
bitter_df["Predicted_Taste"] = np.where(bitter_df["Predicted_Taste"] == "Sweet", "Other", bitter_df["Predicted_Taste"])
bitter_df["name"] = np.where(bitter_df["name"] == "Sweet", "Other", bitter_df["name"])
```

```python
# calculate sensitivity and specificity for bitter and sweet
from sklearn.metrics import confusion_matrix

def sensitivity_specificity(y_true, y_pred):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)
    return sensitivity, specificity

# calculate sensitivity and specificity for bitter
se_bitter, sp_bitter = sensitivity_specificity(bitter_df["name"], bitter_df["Predicted_Taste"])

# calculate sensitivity and specificity for sweet
se_sweet, sp_sweet = sensitivity_specificity(sweet_df["name"], sweet_df["Predicted_Taste"])

# print
print("Sweet sensitivity: ", se_sweet.round(4)*100)
print("Sweet specificity: ", sp_sweet.round(4)*100)
print("Bitter sensitivity: ", se_bitter.round(4)*100)
print("Bitter specificity: ", sp_bitter.round(4)*100)
```
